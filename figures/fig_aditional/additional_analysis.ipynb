```python
import seaborn as sns

with use_style("pyloric"):
    fig,axes = plt.subplots(3,6, figsize=(8,4))

    Ts = [2,11,101]
    tasks  = ["periodic_sde", "general_sde", "mixture_rw_5d", "double_well"]

    axes[0,0].set_title("Periodic SDE")
    axes[0,1].set_title("Linear SDE")
    axes[0,2].set_title("Mixture RW (5d)")
    axes[0,3].set_title("Double well")
    axes[0,4].set_title("Lotka Volterra")
    axes[0,5].set_title("SIR")

    for i in range(3):
        for j in range(6):
            t = Ts[i]
            if j < 4:
                task = tasks[j]
                df1 = query("results/bm_synthetic_10k", metric="swd", method=["nle", "nre", "npe",  "diffusion"], eval_num_steps=t,  num_simulations=[1_000, 10_000],task=task)
                df2 = query("results/bm_synthetic_100k", metric="swd", method=["nle", "nre", "npe",  "diffusion"], eval_num_steps=t, num_simulations=[10_000, 100_000], task=task)
                df3 = query("results/bm_synthetic_1000k", metric="swd", method=["nle", "nre", "npe",  "diffusion"], eval_num_steps=t, task=task)

                df1["num_sims"] = 1000.
                df2["num_sims"] = 10000.
                df3["num_sims"] = 100000.



                df = pd.concat([df1,df2,df3]).reset_index()
            elif j==4:
                df1_lv = query("results/bm_lv100k", metric="swd", method=["nle", "nre", "diffusion"], eval_num_steps=t, num_simulations=[10_000], task="lotka_volterra")
                df1_lv2 = query("results/bm_lv100k", metric="swd", method=["npe"], eval_num_steps=t, num_simulations=[1_000], task="lotka_volterra")
                df1_lv = pd.concat([df1_lv, df1_lv2]).reset_index()

                df2_lv = query("results/bm_lv100k", metric="swd", method=["nle", "nre", "diffusion"], eval_num_steps=t, num_simulations=[100_000], task="lotka_volterra")
                df2_lv2 = query("results/bm_lv100k", metric="swd", method=["npe"], eval_num_steps=t, num_simulations=[10_000], task="lotka_volterra")
                df2_lv = pd.concat([df2_lv, df2_lv2]).reset_index()
                df3_lv = query("results/bm_lv1000k", metric="swd", method=["nle", "nre", "npe",  "diffusion"], eval_num_steps=t, task="lotka_volterra")

                df1_lv["num_sims"] = 1000.
                df2_lv["num_sims"] = 10000.
                df3_lv["num_sims"] = 100000.

                df = pd.concat([df1_lv,df2_lv,df3_lv]).reset_index()
            elif j==5:

                df1_sir = query("results/bm_sir10k", metric="swd", method=["nle", "nre", "diffusion", "npe"], eval_num_steps=t, num_simulations=[10_000, 1000], task="sir")
                df2_sir = query("results/bm_sir100k", metric="swd", method=["nle", "nre", "diffusion","npe"], eval_num_steps=t, num_simulations=[100_000, 10_000], task="sir")
                df3_sir = query("results/bm_sir1000k", metric="swd", method=["nle", "nre", "npe",  "diffusion"], eval_num_steps=t, task="sir")

                df1_sir["num_sims"] = 1000.
                df2_sir["num_sims"] = 10000.
                df3_sir["num_sims"] = 100000.

                df = pd.concat([df1_sir,df2_sir,df3_sir]).reset_index()


            sns.pointplot(df,x="num_sims", y="value", hue="method", palette=DEFAULT_COLORS, ax=axes[i,j], legend=False, markers=".",dodge=False,
        alpha=0.7,
        markersize=2.,
        lw=2.)
            axes[i,j].set_yscale('log', base=2)
            axes[i,j].set_xticklabels(["$10^4$", "$10^5$", "$10^6$"])
            axes[-1, j].set_xlabel("Training budget")
            axes[i, 0].set_ylabel(f"sW [T={t-1}]")
            axes[i,j].set_ylim(0,2**3)

            if j > 0:
                axes[i,j].set_ylabel(None)

            if i < 2:
                axes[i,j].set_xlabel(None)

            plt.tight_layout()

        fig.savefig("plot_by_num_sims.svg")
plt.show()
```
